```python
# Histogram of numerical columns
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

plt.figure(figsize=(16,10))
for i, col in enumerate(num_cols):
    plt.subplot(2,2,i+1)
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Boxplot for outliers
plt.figure(figsize=(16,6))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot of Numerical Features")
plt.show()
```

Sales & Profit Distribution: Right-skewed
Discount Impact: Most orders have small/no discoun
Order Size: Typical orders are small (1–4 items), bulk orders are rare

Sales & Profit have significant outliers.

Very high sales → few large orders dominate revenue.

Very high or negative profit → some orders are very profitable, others incur large losses.

```python
cat_cols = ['Segment', 'Ship Mode', 'Category', 'Sub-Category', 'Region']

plt.figure(figsize=(16,12))
for i, col in enumerate(cat_cols):
    plt.subplot(3,2,i+1)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f'Count of {col}')
plt.tight_layout()
plt.show()

```

Customer Segment

Top Segments: Consumer and Corporate dominate the orders.

Home Office contributes the least.

Insight: Focus marketing or high-value campaigns on Consumer and Corporate segments; Home Office has less volume but may include high-value orders.

Region Analysis

Dominant: West → highest sales

Following: East

Lowest: South → lowest sales

Insight: Regional focus for promotions or logistics can optimize revenue. West and East are strong markets.

Ship Mode

Dominant: Standard Class

Lowest: Same Day

Insight: Most orders follow standard shipping. Same day delivery is rare but could indicate high-cost, low-volume service.

```python
plt.figure(figsize=(10,6))
corr = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Correlation Insights

Sales ↔ Profit: Strongest correlation → higher sales generally mean higher profit.

Quantity ↔ Discount: Weak negative correlation (~-0.22) → increasing discount slightly reduces quantity/profit?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Segment')
plt.title("Profit vs Discount by Segment")
plt.show()
```

Profit vs Discount by Segment:

0% discount → very high profits (up to 8,000)

70–80% discount → huge losses (0 to -4,000)

Insight: Excessive discounts erode profitability. Small or zero discount often yields high-profit orders.

```python

category_sales = df.groupby('Category')[['Sales','Profit']].sum().reset_index()
category_sales_melted = category_sales.melt(
    id_vars='Category', 
    value_vars=['Sales','Profit'], 
    var_name='Metric', 
    value_name='Amount'
)

sns.barplot(data=category_sales_melted, x='Category', y='Amount', hue='Metric')
plt.title("Total Sales & Profit by Category")
plt.show()

```

Category Analysis

Major Category: Office Supplies (largest share of orders).

Furniture: Second highest sales (~700k–800k) but lowest profit → margin issues.

Technology: Highest overall sales and profit → highly profitable category.

Insight:

Furniture: Sales are high but profit is low → investigate discounting or cost structure.

Office Supplies: Lower sales than furniture but higher profit → efficient sales strategy.

```python
sub_sales = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values('Sales', ascending=False)
sub_sales.plot(kind='bar', figsize=(14,6))
plt.title("Sales & Profit by Sub-Category")
plt.show()

```

Sub-Category Analysis

High Sales: Phones, Chairs, Tables

High Profit: Copiers, Paper

Observations:

Copiers → medium sales but highest profit → high margin products.

Paper → low sales but very high profit → niche high-margin product.

Phones → high sales, second-highest profit.

Chairs → similar sales as phones but lower profit → lower margin.

Tables → high sales but negative profit → high loss potential.

Machines → medium sales, minor profit → low contribution.

Insight: Focus on Copiers and Paper for profitability and review Tables for pricing or discount issues.
